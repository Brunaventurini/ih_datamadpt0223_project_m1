```python
#Using DuckDB
conn = duckdb.connect("./data/bicimad_data.db",read_only = True)

df2 = conn.execute("SELECT * FROM bicimad_stations").fetch_df()
#def cleanTxt(text):
    #if text[0] == "'":
        #text = re.sub(r"'", '', '') #Removing empty categories   
    #if text == "]":
        #text = re.sub(r"", '', '') #Removing empty categories  
    #if text == "'":
        #text = re.sub(r"", '', '') #Removing empty categories  
    #return text
df2["geometry.coordinates"] = df2["geometry.coordinates"].str.replace('[', '').str.replace(']', '')
# dropping null value columns to avoid errors :0       
df2.dropna(inplace = True)
 
# new data frame with split value columns
new = df2["geometry.coordinates"].str.split(",", expand = True)
 
# making separate first name column from new data frame
df2["long_finish"]= new[0].apply(lambda x: x.strip())#.apply(lambda x: x.replace("0       ", "")).apply(lambda x: x.replace("0        ", ""))
df2['long_finish'] = df2['long_finish'].astype(float)

# making separate last name column from new data frame
df2["lat_finish"]= new[1].apply(lambda x: x.strip())#.apply(lambda x: x.replace("0       ", "")).apply(lambda x: x.replace("0        ", ""))
df2['lat_finish'] = df2['lat_finish'].astype(float)
# Dropping old columns
df2.drop(columns =["geometry.coordinates"], inplace = True)
 
# df display
df2
```

```python
#results = [[]]
#bici_address = []
result_final = []
station_name_final = []
station_address_final = []
start_name_final = []
start_address_final = []
b = 0
for i in range(len(new_data_frame)):
    place1 = (new_data_frame["Lat_start"][i],new_data_frame["Long_start"][i])
    place1_name = new_data_frame["Place Of Interest"][i]
    place1_address = new_data_frame["Place Address"][i]
    results = []
    station_name = []
    station_address = []
    start_name = []
    start_address = []
    for j in range(len(df2)):
        place2 = (df2["lat_finish"][j],df2['long_finish'][j])
        place2_name = df2["name"][j]
        place2_address = df2["address"][j]
        results.append(distance.distance(place1,place2).km)
        station_name.append(place2_name)
        station_address.append(place2_address)    
        start_name.append(place1_name)
        start_address.append(place1_address)
        #b =+ 1 
        #min_value = min(results)
    result_final.append([results])
    station_name_final.append([station_name])
    station_address_final.append([station_address])
    start_name_final.append([start_name])
    start_address_final.append([start_address])
    #result_final.append(min_value)
```
